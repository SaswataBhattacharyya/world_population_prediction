```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick

population_cols = ['Country/Territory', 'Rank', '2022 Population', '2020 Population', 
                   '2015 Population', '2010 Population', '2000 Population', 
                   '1990 Population', '1980 Population', '1970 Population']

df_pop = df[population_cols]

# Filter top 10 ranked countries
top_10 = df_pop[df_pop['Rank'] <= 10].copy()

# Convert wide data to long format for easier plotting
df_long = top_10.melt(id_vars=['Country/Territory', 'Rank'], 
                       var_name='Year', 
                       value_name='Population')

# Convert 'Year' column to integer
df_long['Year'] = df_long['Year'].str.extract('(\d+)').astype(int)

# Sort by Year
df_long = df_long.sort_values(by=['Year', 'Rank'])

fig, ax = plt.subplots(figsize=(12, 6))  # Explicitly define figure and axis
sns.lineplot(data=df_long, x='Year', y='Population', hue='Country/Territory', marker='o', linewidth=2.5, ax=ax)

ax.set_title('Top 10 Countries - Population Growth Over Decades', fontsize=14)
ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Population (in Billions)', fontsize=12)

# Format y-axis to show population in Billions (B) or Millions (M)
ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{x/1e9:.1f}B' if x >= 1e9 else f'{x/1e6:.0f}M'))

ax.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
ax.grid(True)

plt.show()
```

```python
growth_rate_cols = ['Country/Territory', 'Growth Rate']

# Sort by Growth Rate in descending order
top_10_growth = df[growth_rate_cols].sort_values(by='Growth Rate', ascending=False).head(10)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=top_10_growth, x='Growth Rate', y='Country/Territory', hue='Country/Territory', palette='viridis')

plt.title('Top 10 Countries by Population Growth Rate', fontsize=14)
plt.xlabel('Growth Rate (%)', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```

```python
# Select relevant columns for growth rate
growth_rate_cols = ['Country/Territory', 'Growth Rate']

# Sort by Growth Rate in descending order
top_10_growth = df[growth_rate_cols].sort_values(by='Growth Rate', ascending=False).head(10)

# Extract population columns
population_cols = ['Country/Territory', '2022 Population', '2020 Population', 
                   '2015 Population', '2010 Population', '2000 Population', 
                   '1990 Population', '1980 Population', '1970 Population']

# Filter data for the top 10 fastest-growing countries
df_top_growth = df[df['Country/Territory'].isin(top_10_growth['Country/Territory'])][population_cols]

# Convert wide data to long format
df_long_growth = df_top_growth.melt(id_vars=['Country/Territory'], 
                                    var_name='Year', 
                                    value_name='Population')

# Convert 'Year' column to integer
df_long_growth['Year'] = df_long_growth['Year'].str.extract('(\d+)').astype(int)

# Sort by Year
df_long_growth = df_long_growth.sort_values(by=['Year'])

# Plot
fig, ax = plt.subplots(figsize=(12, 6))
sns.lineplot(data=df_long_growth, x='Year', y='Population', hue='Country/Territory', marker='o', linewidth=2.5, ax=ax)

ax.set_title('Population Growth of Top 10 Fastest-Growing Countries', fontsize=14)
ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Population (in Billions)', fontsize=12)

# Format y-axis to display Billions (B) or Millions (M)
import matplotlib.ticker as mtick
ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{x/1e9:.1f}B' if x >= 1e9 else f'{x/1e6:.0f}M'))

ax.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
ax.grid(True)

plt.show()
```

```python
# Sort by 2022 Population and take top 10
top_10_countries = df.nlargest(10, '2022 Population')

plt.figure(figsize=(12, 6))
ax = sns.barplot(x='2022 Population', y='Country/Territory', hue='Country/Territory', data=top_10_countries, palette='viridis')
plt.xlabel('Population in 2022')
plt.ylabel('Country')
plt.title('Top 10 Most Populated Countries in 2022')

# Format y-axis to show population in Billions or Millions
ax.xaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{x/1e9:.1f}B' if x >= 1e9 else f'{x/1e6:.0f}M'))

plt.show()
```

```python
# Group by Continent and sum the population
continent_population = df.groupby('Continent')['2022 Population'].sum().reset_index()

# Plot continent-wise population with hue for continent
plt.figure(figsize=(12, 6))
ax = sns.barplot(x='Continent', y='2022 Population', hue='Continent', data=continent_population, palette='magma')
plt.xlabel('Continent')
plt.ylabel('Total Population in 2022')
plt.title('Population by Continent in 2022')

# Format y-axis to show population in Billions or Millions
ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{x/1e9:.1f}B' if x >= 1e9 else f'{x/1e6:.0f}M'))

plt.xticks(rotation=45)
plt.show()
```

```python
# Sort and filter the top 10 countries by Area
top_10_countries_map = df[['Country/Territory', 'Area (km²)']].sort_values(by='Area (km²)', ascending=False).head(10)
top_10_countries_map = top_10_countries_map.rename(columns={'Country/Territory': 'Country'})

# Plot the bar chart with color gradient
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Area (km²)', data=top_10_countries_map, palette='viridis')
plt.xlabel('Country')
plt.ylabel('Area (km²)')
plt.title('Top 10 Countries by Area (km²)')

# Add value labels to each bar
for index, value in enumerate(top_10_countries_map['Area (km²)']):
    plt.text(index, value, f'{value:,.0f}', ha='center', va='bottom')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from geodatasets import get_path

shapefile_path = '/home/saswata/data_science/UM_projects/world_population/data/10m_cultural/10m_cultural/ne_10m_admin_0_countries.shp'

# Load world map from geodatasets
world = gpd.read_file(shapefile_path)

top_10_countries_map = df[['Country/Territory', 'Area (km²)']].sort_values(by='Area (km²)', ascending=False).head(10)
top_10_countries_map = top_10_countries_map.rename(columns={'Country/Territory': 'ADMIN'})

world_top10 = world.merge(top_10_countries_map, on='ADMIN', how='inner')

# Plot
fig, ax = plt.subplots(figsize=(12, 8))
world_top10.plot(column='Area (km²)', ax=ax, legend=True, cmap='viridis',
                 legend_kwds={'label': "Area in km²"})
ax.set_title('Top 10 Countries by Area (km²)', fontsize=15)
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# Assuming df is already defined with country and area data
df_countries = df[['Country/Territory', 'Area (km²)']].copy()
df_countries = df_countries.rename(columns={'Country/Territory': 'ADMIN'})

# Merge with shapefile GeoDataFrame
world_merged = world.merge(df_countries, on='ADMIN', how='left')

# Plot full map
fig, ax = plt.subplots(figsize=(15, 10))
world_merged.plot(column='Area (km²)', ax=ax, legend=True, cmap='viridis',
                  legend_kwds={'label': "Area in km²"})
ax.set_title('Countries by Area (km²)', fontsize=15)
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
# Sort and filter the top 10 countries by Density
top_10_density_map = df[['Country/Territory', 'Density (per km²)']].sort_values(by='Density (per km²)', ascending=False).head(10)
top_10_density_map = top_10_density_map.rename(columns={'Country/Territory': 'Country'})

# Plot the bar chart with color gradient
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Density (per km²)', data=top_10_density_map, palette='viridis')
plt.xlabel('Country')
plt.ylabel('Density (per km²)')
plt.title('Top 10 Countries by Population Density')

# Add value labels to each bar
for index, value in enumerate(top_10_density_map['Density (per km²)']):
    plt.text(index, value, f'{value:,.0f}', ha='center', va='bottom')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
